#**Data Preprocessing**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_std = df[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']]
```

```python
x_std = sc.fit_transform(x_std)
```

```python
x[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']] = pd.DataFrame(x_std,columns=[['Item_Weight','Item_Visibility','Item_MRP','Outlet_Establishment_Year']])
```

#**Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.1,random_state = 2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

#**Modeling**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(random_state=2529)
```

```python
y_train = y_train.astype(str)
```

```python
x_train = pd.get_dummies(x_train)
```

```python
# Impute missing values using SimpleImputer
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor

# Assuming x_train is your DataFrame with missing values
imputer = SimpleImputer(strategy='mean')  # Or another strategy like 'median', 'most_frequent'
x_train_imputed = imputer.fit_transform(x_train)

# Now fit the model with imputed data
rfr = RandomForestRegressor(random_state=2529)
rfr.fit(x_train_imputed, y_train)
```

#**Prediction**

```python
x_test_imputed = imputer.transform(x_test)
```

```python
y_pred = rfr.predict(x_test_imputed)
```

```python
y_pred.shape
```

```python
y_pred
```

#**Evaluation**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
